```python
random_search = RandomizedSearchCV(clf,param_distributions=param_dict,n_iter=n_iter_search,scoring="roc_auc",cv=10)
```

```python
random_search.fit(x_train,y_train)
```

```python
def report(results, n_top=3):
    for i in range(1,n_top+1):
        candidates = np.flatnonzero(results['rank_test_score']==i)
        for candidate in candidates:
            print("Model with rank:{0}".format(i))
            print("Mean variation score:{0:3f} (std:{1:3f})".format(results['mean_test_score'][candidate],
                                                                   results['std_test_score'][candidate]))
            print("Parameters:{0}".format(results['params'][candidate]))
```

```python
np.flatnonzero(random_search.cv_results_['rank_test_score']==1)
```

```python
report(random_search.cv_results_,5)
```
